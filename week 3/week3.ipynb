# Import Libraries

```python
# Import required libraries

import pandas as pd
import sqlite3
```

### Data Loading and Exploration

**Libraries Imported**: The necessary libraries, `pandas` for data manipulation and `sqlite3` for database interaction, were successfully imported.

# Load Dataset

```python
# Load Excel dataset

df = pd.read_excel("Cleaned_Dataset_Project1.xlsx")

# Display first rows

df.head()
```

**Dataset Loading**: An Excel file named `Cleaned_Dataset_Project1.xlsx` was loaded into a pandas DataFrame (`df`). The `df.head()` output displayed the first few rows, showing columns like `OrderID`, `Date`, `Product`, `TotalPrice`, `PaymentMethod`, `OrderStatus`, and `ReferralSource`.

# Explore Dataset

```python
# Dataset information

print("Dataset Shape:")
print(df.shape)

print("\nColumns:")
print(df.columns)

print("\nMissing Values:")
print(df.isnull().sum())
```

**Dataset Information**: The dataset contains `1200` rows and `15` columns. All columns appear to be complete with no missing values, which is excellent for analysis. The column names indicate a sales transaction dataset.

# Create SQL Database

```python
# Create an in-memory SQLite database

conn = sqlite3.connect(':memory:')

# Load dataframe into SQL table

df.to_sql(
    'sales_data',
    conn,
    if_exists='replace',
    index=False
)

print("Table successfully created.")
```

### SQL Database Creation

**In-memory SQLite Database**: An in-memory SQLite database was created, and your `df` DataFrame was successfully loaded into a table named `sales_data`. This allows for SQL-based querying of your dataset.

# Basic SELECT Queries

Query 1 - View All Records



```python
query = """
SELECT *
FROM sales_data
LIMIT 10;
"""

pd.read_sql(query, conn)
```

### Basic SELECT Queries

**View All Records (LIMIT 10)**: The query `SELECT * FROM sales_data LIMIT 10;` successfully retrieved the first 10 rows of the `sales_data` table, providing a quick overview of the data structure and content within the SQL environment.

# Select Specific Columns

```python
query = """
SELECT
    OrderID,
    Product,
    Quantity,
    TotalPrice
FROM sales_data
LIMIT 10;
"""

pd.read_sql(query, conn)
```

**Select Specific Columns**: The query `SELECT OrderID, Product, Quantity, TotalPrice FROM sales_data LIMIT 10;` demonstrated how to select a subset of columns, showing order details such as product and total price for the first 10 records.

# WHERE Clause

Delivered Orders

```python
query = """
SELECT *
FROM sales_data
WHERE OrderStatus = 'Delivered';
"""

pd.read_sql(query, conn)
```

### WHERE Clause Analysis

**Delivered Orders**: A query filtered the data for orders with `OrderStatus = 'Delivered'`. The output shows there are `231` delivered orders, providing insight into the proportion of successful deliveries.

High Value Orders

```python
query = """
SELECT *
FROM sales_data
WHERE TotalPrice > 2000;
"""

pd.read_sql(query, conn)
```

**High Value Orders**: A query filtered for orders where `TotalPrice > 2000`. The result identified `180` high-value orders, indicating a segment of customers or products contributing significantly to revenue.

# ORDER BY

Highest Sales

```python
query = """
SELECT
    OrderID,
    Product,
    TotalPrice
FROM sales_data
ORDER BY TotalPrice DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

### ORDER BY Analysis

**Highest Sales**: The query `ORDER BY TotalPrice DESC LIMIT 10;` successfully identified the top 10 highest sales transactions. The product `Tablet` had the single highest sale at `3456.40`, followed closely by `Monitor` and `Laptop` sales.

Lowest Sales

```python
query = """
SELECT
    OrderID,
    Product,
    TotalPrice
FROM sales_data
ORDER BY TotalPrice ASC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

**Lowest Sales**: Conversely, the query `ORDER BY TotalPrice ASC LIMIT 10;` showed the 10 lowest sales. The lowest recorded sale was for a `Phone` at `11.39`, indicating a wide range in transaction values.

# COUNT Aggregation

Total Number of Orders

```python
query = """
SELECT COUNT(*) AS TotalOrders
FROM sales_data;
"""

pd.read_sql(query, conn)
```

### COUNT Aggregation

**Total Number of Orders**: The `COUNT(*)` query confirmed that there are `1200` total orders in the `sales_data` table.

Orders by Status

```python
query = """
SELECT
    OrderStatus,
    COUNT(*) AS NumberOfOrders
FROM sales_data
GROUP BY OrderStatus
ORDER BY NumberOfOrders DESC;
"""

pd.read_sql(query, conn)
```

**Orders by Status**: Grouping orders by `OrderStatus` revealed the distribution: `Cancelled` (250), `Returned` (247), `Pending` (237), `Shipped` (235), and `Delivered` (231). This shows that `Cancelled` orders are the most frequent, which could be an area for investigation.

# SUM Aggregation

Total Revenue

```python
query = """
SELECT
    SUM(TotalPrice) AS TotalRevenue
FROM sales_data;
"""

pd.read_sql(query, conn)
```

### SUM Aggregation

**Total Revenue**: The sum of `TotalPrice` across all orders revealed a total revenue of `1,264,761.96` from the sales data.

Revenue by Product

```python
query = """
SELECT
    Product,
    SUM(TotalPrice) AS Revenue
FROM sales_data
GROUP BY Product
ORDER BY Revenue DESC;
"""

pd.read_sql(query, conn)
```

**Revenue by Product**: This analysis, `GROUP BY Product ORDER BY Revenue DESC;`, showed that `Chair` is the top revenue-generating product with `195,620.11`, closely followed by `Printer` and `Laptop`. `Phone` generated the least revenue at `151,722.39`.

# AVG Aggregation

Average Order Value

```python
query = """
SELECT
    AVG(TotalPrice) AS AverageOrderValue
FROM sales_data;
"""

pd.read_sql(query, conn)
```

### AVG Aggregation

**Average Order Value**: The overall average order value across all transactions is `1053.97`.

Average Revenue by Product

```python
query = """
SELECT
    Product,
    AVG(TotalPrice) AS AverageRevenue
FROM sales_data
GROUP BY Product
ORDER BY AverageRevenue DESC;
"""

pd.read_sql(query, conn)
```

**Average Revenue by Product**: When looking at average revenue per product (`GROUP BY Product ORDER BY AverageRevenue DESC;`), `Laptop` leads with an average of `1110.56`, followed by `Chair` and `Printer`. `Phone` also has the lowest average revenue per order among products.

# Advanced GROUP BY Analysis

Payment Method Performance

```python
query = """
SELECT
    PaymentMethod,
    COUNT(*) AS Orders,
    SUM(TotalPrice) AS Revenue,
    AVG(TotalPrice) AS AverageOrderValue
FROM sales_data
GROUP BY PaymentMethod
ORDER BY Revenue DESC;
"""

pd.read_sql(query, conn)
```

### Advanced GROUP BY Analysis

**Payment Method Performance**: This query grouped sales data by `PaymentMethod`. `Credit Card` is the highest revenue generator (`263,847.63`) and also has the highest average order value (`1127.55`). While `Online` payments have the most orders (`258`), their average order value is slightly lower.

Referral Source Performance

```python
query = """
SELECT
    ReferralSource,
    COUNT(*) AS Orders,
    SUM(TotalPrice) AS Revenue
FROM sales_data
GROUP BY ReferralSource
ORDER BY Revenue DESC;
"""

pd.read_sql(query, conn)
```

**Referral Source Performance**: Grouping by `ReferralSource` shows that `Instagram` is the leading source in terms of revenue (`275,285.45`) and also has the most orders (`259`). `Email` and `Google` are also significant contributors.
