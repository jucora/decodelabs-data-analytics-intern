**Import Libraries and Load the Dataset**

This step imports the Pandas library and loads the dataset into a DataFrame. The head() method is used to preview the first records and verify that the data has been loaded correctly.

```python
import pandas as pd

# Load the Excel file
df = pd.read_excel("Dataset for Data Analytics.xlsx")

# Display the first rows
print(df.head())
```

**Dataset Overview**

This step provides a general overview of the dataset, including:


*   Number of rows and columns
*   Data types
*   Missing values

This is usually the first step in any data cleaning process.

```python
# Number of rows and columns
print(df.shape)

# Data types and null values
print(df.info())
```

**Missing Values Analysis**

The isnull() function identifies missing values, while sum() counts them.

**CouponCode**: 309

All other columns should contain: 0

```python
# Count missing values in each column
missing_values = df.isnull().sum()

print(missing_values)
```

**Handle Missing Values**

The fillna() method replaces missing values.

Using "No Coupon" clearly indicates that the customer did not apply any coupon during the purchase.

```python
# Replace missing values in CouponCode
df["CouponCode"] = df["CouponCode"].fillna("No Coupon")
```

**Verify Missing Values Again**

This step confirms that all missing values were successfully handled.

**CouponCode**: 0

```python
print(df.isnull().sum())
```

**Check for Duplicate Rows**

The duplicated() method identifies repeated records.

**Duplicate Rows**: 0

```python
# Count duplicate rows
duplicate_rows = df.duplicated().sum()

print("Duplicate Rows:", duplicate_rows)
```

**Check for Duplicate Order IDs**

Since **OrderID** should be unique, we verify that no identifier appears more than once.

**Duplicate Order IDs**: 0

```python
duplicate_order_ids = df["OrderID"].duplicated().sum()

print("Duplicate Order IDs:", duplicate_order_ids)
```

**Validate Date Format**

pd.to_datetime() converts values into proper datetime objects.

The parameter:

errors="coerce"

converts invalid dates into NaT (Not a Time).

If there are no invalid dates:

Invalid Dates: 0

```python
# Convert Date column to datetime
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

# Check invalid dates
invalid_dates = df["Date"].isnull().sum()

print("Invalid Dates:", invalid_dates)
```

**Standardize Date Format**

This ensures consistency across all records.

Example: 2025-05-08

```python
# Format dates as YYYY-MM-DD
df["Date"] = df["Date"].dt.strftime("%Y-%m-%d")

print(df["Date"])
```

**Validate Numeric Consistency**

Business rule:

**TotalPrice** = Quantity × UnitPrice

This verifies that all total prices were calculated correctly.

The validation confirmed that all records satisfy the pricing rule after rounding values to two decimal places.

This indicates that the `TotalPrice` values stored in the dataset are consistent with the expected calculation based on `Quantity` and `UnitPrice`. Any discrepancies observed during the initial validation were caused by floating-point precision differences rather than actual data quality issues.

**Price Inconsistencies**: 0

```python
df["CalculatedTotal"] = (
    df["Quantity"] * df["UnitPrice"]
).round(2)

invalid_prices = (
    df["TotalPrice"].round(2) !=
    df["CalculatedTotal"]
).sum()

print("Price Inconsistencies:", invalid_prices)
```

**Review Categorical Columns**

The unique() method displays all distinct values.

This helps detect issues such as: Shipped, shipped, SHIPPED

No inconsistencies were found.

```python
print(df["PaymentMethod"].unique())
print(df["OrderStatus"].unique())
print(df["ReferralSource"].unique())
```

**Final Validation Summary**

This generates a concise report summarizing the dataset quality after cleaning.

```python
print("========== DATA CLEANING SUMMARY ==========")
print("Missing Values:", df.isnull().sum().sum())
print("Duplicate Rows:", df.duplicated().sum())
print("Duplicate Order IDs:", df["OrderID"].duplicated().sum())
print("Total Records:", len(df))
```

**Save the Cleaned Dataset**

This creates the final deliverable required by the project.

**Output**: Cleaned_Dataset_Project1.xlsx

```python
# Remove temporary column
#df.drop(columns=["CalculatedTotal"], inplace=True)

# Save cleaned dataset
df.to_excel(
    "Cleaned_Dataset_Project1.xlsx",
    index=False
)

print("Dataset successfully saved.")
```
